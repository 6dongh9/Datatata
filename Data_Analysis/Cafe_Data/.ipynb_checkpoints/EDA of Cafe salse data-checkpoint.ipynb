# 스터디 카페 데이터를 활용한 EDA

```python
# 라이브러리 호출
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

```python
# 한글 시각화를 위한 seaborn 폰트설정
plt.rcParams['axes.unicode_minus'] = False
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(context='notebook',
              style='darkgrid',
              palette='pastel',
              font='NanumBarunGothic',
              rc=custom_params)
```

```python
# 데이터 로드
entry_data = pd.read_csv('/aiffel/data/entry_data.csv')
sales_data = pd.read_csv('/aiffel/data/sales_data.csv')
item_data = pd.read_csv('/aiffel/data/item_data.csv')
item_entry_data = pd.read_csv('/aiffel/data/item_entry_data.csv')
```

### 1. EDA
**A. 월별 매출액 파악**

```python
temp_var = sales_data # 활용할 데이터
temp_var['date'] = sales_data["결제일시"].apply(lambda x: x[:7]) # date 컬럼 생성
```

```python
temp_var["결제금액"] = temp_var["결제금액"].str.replace(",","") # 결제금액에 포함된 쉼표(,) 제거
temp_var['결제금액'] = temp_var['결제금액'].astype('int') # 자료형 변경
payment_data = temp_var.groupby('date')['결제금액'].sum().reset_index() # date 컬럼 기준으로 payment 집계
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=payment_data, x='date', y='결제금액')
plt.xlabel('')
plt.ylabel('금액 (천만 원)')
plt.xticks(rotation=45)
plt.title('월별 매출액')
plt.show()
```

**B. 인기 상품 파악**

```python
temp_var = temp_var[temp_var['판매수량'] > -1] # 결제취소 건 제외
x = temp_var['상품명'].value_counts()

# 집계 결과를 데이터프레임으로 변환하고 상위 10개 상품 선정
items_count = pd.DataFrame(x)
temp_item = items_count.head(10)
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=temp_item, x=temp_item.index, y='상품명')
plt.xlabel('상품명')
plt.ylabel('판매량')
plt.xticks(rotation=45)
plt.title('상품 판매 수량 TOP 10')
plt.show()
```

```python
items_payment = temp_var.groupby('상품명')['결제금액'].sum().reset_index() # 결제금액 기준 집계

# 집계 결과를 데이터프레임으로 변환하고 상위 10개 상품 선정
items_payment = items_payment.sort_values(by='결제금액', ascending=False).reset_index(drop=True)
temp_payment = items_payment.head(10)
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=temp_payment, x='상품명', y='결제금액')
plt.xlabel('상품명')
plt.ylabel('금액 (백만 원)')
plt.xticks(rotation=45)
plt.title('상품 매출액 TOP 10')
plt.show()
```

* 필수 문제3: 가장 많이 팔린 입장권 종류(entry_data를 이용합니다.)
* 여러 입장권 종류를 판매 매출(판매 개수 x 금액) 순으로 정렬하고 상위 10개를 뽑아 결과를 시각화해봅니다.
* 힌트: value_counts() 함수를 사용하세요.

**C. 인기 입장권 파악**

```python
temp_var = entry_data # 활용할 데이터

# 전처리
temp_var['금액'] = temp_var['금액'].astype('int') # 자료형 변경
tickets= temp_var.groupby('요금제명')['금액'].sum().reset_index() # 요금제명 기준 집계
temp_tickets = tickets.sort_values(by='금액', ascending=False).reset_index(drop=True).head(10) # 상위 10개 항목 추출
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=temp_tickets, x='요금제명', y='금액')
plt.xlabel('요금제명')
plt.ylabel('금액 (백만 원)')
plt.xticks(rotation=45)
plt.title('입장권 매출액 TOP 10')
plt.show()
```

### 2. 문제 정의
**A. 시간대별 매출액 파악**
1. Problem: 매출이 적은 시간대를 파악하고 매출 증진 전략을 수립할 필요가 있음  

2. Solution: 시간대별 매출액을 집계해 확인한 후 대안 모색

```python
temp_var = sales_data # 활용할 데이터
temp_var['time'] = sales_data["결제일시"].apply(lambda x: x[11:13]) # 결제 시간 컬럼 생성
temp_var = temp_var[temp_var['판매수량'] > -1] # 결제취소 건 제외
time_sum = temp_var.groupby('time')['결제금액'].sum().reset_index() # time 컬럼 기준 결제금액 합계 계산
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=time_sum, x='time', y='결제금액')
plt.xlabel('시간')
plt.ylabel('금액 (백만 원)')
plt.title('시간대별 매출액')
plt.show()
```

* 시간대별 확인 결과 저녁 시간대에 고객이 가장 많이 몰리는 것을 확인할 수 있음
* 20시 이후 매출이 크게 감소하므로 20시~24시 사이에는 음료 또는 디저트에 대해 프로모션을 진행하여 해당 시간대 매출 증가를 기대할 수 있음

**B. 카테고리별 매출액 파악**
1. Problem: 매출 상위 카테고리 현황 파악 필요
2. Solution: 고객이 선호하는 카테고리 파악 후 매출 증진 전략 수립

```python
temp_data = sales_data[sales_data["판매수량"] > -1] # 결제취소 건 제외
temp_data = pd.merge(temp_data, item_data, on="상품명") # 상품명 기준 데이터 병합
print('주의: 판매금액과 결제금액 의 차이를 확실하게 구분해야 합니다. (결제금액 = 판매금액 * 판매수량)')

category = temp_data.groupby('분류')['결제금액'].sum().reset_index() # 카테고리 기준 결제금액 집계
category = category.sort_values(by='결제금액', ascending=False).reset_index(drop=True)
```

```python
# 집계 결과 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=category, x='분류', y='결제금액')
plt.xlabel('')
plt.ylabel('금액 (천만 원)')
plt.xticks(rotation=45)
plt.title('카테고리별 매출액')
plt.show()
```

* 음식 카테고리가 매출의 90% 정도를 차지함
* 카페와 맞지 않는 카테고리가 있으나 해당 카테고리의 매출액은 미미한 수준
* 효과적인 매출액 증진을 위해서는 음료 카테고리 개선이 필요함

**C. 비주류 상품 파악**
* Problem: 20시 이후 매출 증진 방안이 필요함
* Solution: 20시 이후 판매되는 상품 현황 파악 및 대안 모색

```python
# 20시 이후 판매되는 카테고리 외 항목 제외
none_meal = temp_data[temp_data["분류"] != '음식']
none_meal = none_meal[none_meal["분류"] != '포장']
none_meal = none_meal[none_meal["분류"] != '게임&TV']

drink = none_meal.groupby(['time','분류'])['결제금액'].sum().reset_index() # 시간대 및 카테고리 기준 결제금액 집계
```

```python
# 집계 결과 시각화
plt.figure(figsize=(15, 6))
sns.barplot(data=drink, x='time', y='결제금액', hue='분류')
plt.xlabel('시간')
plt.ylabel('금액 (원)')
plt.title('시간대/카테고리별 매출액')
plt.show()
```

* 16시에 커피 판매량이 줄고 디저트 판매량이 올라간 것으로 보아 해당 시간대 디저트 수요가 확실히 존재함
* 20시 이후 커피 판매량이 현저히 감소하지만, 디저트를 포함한 다른 항목의 판매량은 그렇지 않음

**E. 고객 등급 구분 및 등급별 매출 확인**
* Problem: 단골 고객 파악 및 보상 전략 수립 필요성
* Solution: 고객 등급을 구분하고 등급별 매출액 파악 및 보상 전략 모색

```python
# 고객별 요금제 결제 금액 집계
temp_var = entry_data[['고객', '금액']]
entry_customer_payment = temp_var.groupby('고객').agg({'금액':'sum'})
entry_customer_payment.sort_values(by='금액', ascending=False, inplace=True)
entry_customer_payment.reset_index(inplace=True)
entry_customer_payment.drop(entry_customer_payment[entry_customer_payment['금액'] == 0].index, inplace=True) # 결제금액 0인 데이터 제거
entry_customer_payment.rename(columns={'고객':'Customer_ID', '금액':'Entry_payment'}, inplace=True) # 컬럼명 변경
print('참고: 집계된 금액이 0인 데이터 행 38개는 결제 후 환불한 고객이라고 판단하여 데이터에서 제거하였습니다.')
```

```python
# 고객별 음식/음료 결제 금액 집계
temp_var = sales_data[['고객', '결제금액']]
# temp_var['결제금액'] = temp_var['결제금액'].str.replace(',', '')
temp_var['결제금액'] = temp_var['결제금액'].astype('int')
item_customer_payment = temp_var.groupby('고객').agg({'결제금액':'sum'})
item_customer_payment.sort_values(by='결제금액', ascending=False, inplace=True)
item_customer_payment.reset_index(inplace=True)
item_customer_payment.drop(item_customer_payment[item_customer_payment['결제금액']==0].index, inplace=True) # 결제금액 0인 데이터 제거
item_customer_payment.rename(columns={'고객':'Customer_ID', '결제금액':'Item_payment'}, inplace=True) # 컬럼명 변경
```

```python
# 데이터 병합
customer_payment = entry_customer_payment.merge(item_customer_payment, on='Customer_ID', how='outer')
customer_payment.fillna(0, inplace=True) # 결측치 0으로 대체
```

###### Try 1) 입장권 결제액, 상품 결제액 기준으로 고객 클러스터링
* K-means 알고리즘 사용

```python
# 데이터 분포 확인
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_payment, x='Entry_payment', y='Item_payment')
```

* 이상치 때문에 제대로 파악이 안됨

```python
# 이상치 분리
customer_outlier = customer_payment[(customer_payment['Entry_payment']>800000)|(customer_payment['Item_payment']>250000)]
customer_payment.drop(customer_payment[(customer_payment['Entry_payment']>800000)|(customer_payment['Item_payment']>250000)].index, inplace=True)

# 이상치를 제거한 데이터 분포 확인
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_payment, x='Entry_payment', y='Item_payment')
```

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing

# K-means Clustering
Kmeans = KMeans(n_clusters=5, n_init=10, random_state=25)
temp_var = customer_payment[['Entry_payment', 'Item_payment']]
Kmeans.fit(temp_var)

# 클러스터링 결과를 새로운 컬럼에 추가
cluster_label = Kmeans.labels_
customer_payment['Cluster'] = pd.DataFrame(cluster_label)
```

```python
# 클러스터링 결과 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_payment, x='Entry_payment', y='Item_payment', hue='Cluster')
plt.ylabel('음식/음료 판매액')
plt.xlabel('입장권 판매액')
```

* 클러스터링 기법은 고객을 구분하기에 적합하지 않음

###### Try 2) 총 결제액으로 고객 등급 구분

```python
customer_payment.append(customer_outlier) # 분리했던 이상치 결합
customer_payment.drop('Cluster', axis=1, inplace=True) # Cluster 컬럼 삭제

# 고객별 총 결제 금액 집계
customer_payment['Total_payment'] = customer_payment['Entry_payment'] + customer_payment['Item_payment'] # Total_payment 컬럼 생성
customer_payment.sort_values(by='Total_payment', ascending=False, inplace=True) # Total_payment 기준 정렬
customer_payment.reset_index(drop=True, inplace=True) # 인덱스 재설정
```

```python
# 등급 구분
customer_payment['Class'] = pd.qcut(customer_payment['Total_payment'], q=[0, 0.4, 0.65, 0.85, 0.99, 1], labels=[5, 4, 3, 2, 1])

# 등급별 시각화
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Pie Chart
temp_var = customer_payment['Customer_ID'].groupby(customer_payment['Class']).count()
ax1.pie(temp_var, labels=temp_var.index, autopct='%1.1f%%')
ax1.legend()
ax1.set_title('Propotion')

# Scatter Plot
ax2 = sns.scatterplot(data=customer_payment, x='Entry_payment', y='Item_payment', hue='Class')
ax2.set_title('Scatter Plot')
plt.ylabel('음식/음료 판매액')
plt.xlabel('입장권 판매액')

plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
# 등급 기준으로 평균 결제액 집계
temp_var = customer_payment.groupby('Class').agg({'Entry_payment':'mean', 'Item_payment':'mean'})
temp_var = temp_var.rename(columns={'Entry_payment':'입장권 판매액', 'Item_payment':'음식/음료 판매액'})

# 등급별 평균 결제액 시각화
temp_var.plot(kind='bar', stacked=True)
plt.show()
```

* 입장권 결제액이 평균 매출액의 대부분을 차지하고 있으므로 입장권 판매량을 늘리는 데 집중해야 함  
* Class 3~5에 해당하는 고객을 대상으로 입장권 및 음식/음료 판매량 촉진을 위한 프로모션을 기획할 수 있음

### 2. 회고

# 게임 내 매출액 감소 원인 분석
### 1. 문제 상황
* 게임 내 매출액 감소

### 2. 활용 데이터
A. sales: 유저 결제 데이터
* install: 게임 설치 일자
* Time_stamp: 최근 접속 일자
* Game_Name: 플레이한 게임명
* Status: 설치 후 플레이 여부

B. customer: 유저 정보 데이터
* Customer_ID: 유저 ID

* Nickname

* MobileCompany: 통신사

```python
# 라이브러리 호출
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### 1. 데이터 준비

```python
# 데이터 로드
sales_data = pd.read_csv('~/aiffel/data_analysis_basic/data/sales_7.csv')
customer_data = pd.read_csv('~/aiffel/data_analysis_basic/data/customer_7.csv')
```

```python
# # 데이터 확인(1)
# sales_data.head()
```

```python
# # 데이터 확인(2)
# customer_data.head()
```

```python
# Customer_ID 컬럼을 기준으로 데이터 병합
customer_03 = pd.merge(sales_data, customer_data, on='Customer_ID')
```

```python
# 4월분 데이터 로드
customer_04 = pd.read_csv('~/aiffel/data_analysis_basic/data/customer_04_7.csv')
```

```python
# 3월 데이터와 4월 데이터 병합
total_sales_data = pd.concat([customer_03, customer_04])
total_sales_data.reset_index(inplace=True)
```

```python
player_data = total_sales_data[total_sales_data['Status']=='play'] # 플레이 중인 고객의 데이터만 추출
player_data['Month'] = player_data['Time_stamp'].str[:7] # Month 컬럼 생성
```

### 2. EDA
**A. 월별 유저 수**

```python
# 월별 유저 수 시각화
ax = sns.countplot(data=player_data, x='Month')

# 컨테이너 생성
containers = ax.containers[0]
ax.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
plt.title('Player Count by Month')
plt.show()
```

**B. 월별 결제 금액**

```python
# 그룹화
grouped = player_data['Payment'].groupby(player_data['Month']).sum()

# 시각화
ax = sns.barplot(data=grouped.reset_index(), x='Month', y='Payment')
containers = ax.containers[0]
ax.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
plt.title('Payment Sum by Month')
plt.show()
```

**C. 게임별 4월 매출액**

```python
# 그룹화
grouped = player_data['Payment'].groupby([player_data['Month'], player_data['Game_Name']]).sum()

# 시각화
ax = sns.barplot(data=grouped.reset_index(), x='Month', y='Payment', hue='Game_Name')
containers0 = ax.containers[0]
containers1 = ax.containers[1]
ax.bar_label(containers0, labels=[f'{x:,.0f}' for x in containers0.datavalues], label_type='center')
ax.bar_label(containers1, labels=[f'{x:,.0f}' for x in containers1.datavalues], label_type='center')
plt.title('Payment Sum by Month')
plt.show()
```

* 두 게임 모두 4월 매출액이 크게 감소함
* MineGround의 매출액 감소가 전체 매출액 감소에 더 큰 영향을 미침  

**D. 게임별, 일자별 결제 금액**

```python
# 4월 데이터 추출
player_data_04 = player_data[player_data['Month']=='2022-04']

# 그룹화
grouped = player_data_04['Payment'].groupby([player_data_04['Time_stamp'], player_data_04['Game_Name']]).sum()

# 시각화
ax = sns.lineplot(data=grouped.unstack())
plt.xticks([0, 10, 20, 28]) # x 레이블에 표시할 값
plt.title('Payment by Game')
plt.show()
```

**E. 게임별, 일자별 유저 수**

```python
# crosstab() 함수로 데이터 집계
temp_var = pd.crosstab(player_data_04['Time_stamp'], player_data_04['Game_Name'])

# 시각화
temp_var.plot()
plt.title('Customer Count by Game')
plt.show()
```

* MineGround의 유저가 대략 4월 11일을 기준으로 크게 감소하므로 이에 대한 원인 파악이 필요함

### 3. Mine Ground 유저 데이터 분석
**A. 성별별 고객 수**

```python
# 데이터 추출
MG_player_data_04 = player_data_04[player_data_04['Game_Name']=='MineGround'].reset_index()

# 데이터 집계
temp_var = pd.crosstab(MG_player_data_04['Time_stamp'], MG_player_data_04['Gender'])

# 시각화
temp_var.plot()
plt.title('Customer Count by Gender')
plt.show()
```

**B. 연령대별 고객 수**

```python
MG_player_data_04['age'] = 2022 - pd.to_numeric(MG_player_data_04['Birth'].str[:4]) # age 컬럼 생성
temp_var = pd.crosstab(MG_player_data_04['Time_stamp'], MG_player_data_04['age']) # 데이터 집계
MG_player_data_04['age'].describe() # 연령대 구분을 위해 기술통계량 확인
```

```python
bins = [16, 20, 30, 40, 50] # 구간 설정
bins_label = ['10', '20', '30', '40'] # 구간 이름 정의
MG_player_data_04['Generation'] = pd.cut(MG_player_data_04['age'], bins, right=False, labels=bins_label) # Generation 컬럼을 생성해서 구간 나누기
temp_var = pd.crosstab(MG_player_data_04['Time_stamp'], MG_player_data_04['Generation']) # 세대에 따른 일자별 고객 수 집계

# 시각화
temp_var.plot()
plt.title('Customer Count by Age')
plt.show()
```

**C. 통신사별 고객 수**

```python
# 데이터 집계
temp_var = pd.crosstab(MG_player_data_04['Time_stamp'], MG_player_data_04['MobileCompany'])

# 시각화
temp_var.plot()
plt.title('Customer Count by Mobile Company')
plt.show()
```

**D. OS별 고객 수**

```python
# 데이터 집계
temp_var = pd.crosstab(MG_player_data_04['Time_stamp'], MG_player_data_04['OS'])

# 시각화
temp_var.plot()
plt.title('Customer Count by OS')
plt.show()
```

* 4월 11일을 기점으로 Android OS 유저의 수가 0명이므로 Android 서버에 문제가 있었음을 알 수 있음

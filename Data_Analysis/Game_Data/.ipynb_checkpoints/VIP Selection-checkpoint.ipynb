# VIP 플레이어 선정
### 1. Problem
* 게임 내 VIP 플레이어에 대한 선정 기준 및 보상 체계가 수립되어 있지 않음  

### 2. Solution
* RFM Score 계산을 통해 VIP 플레이어 선정 및 보상 체계 수립

```python
# 라이브러리 호출
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

### 1. 데이터 준비

```python
# 데이터 로드
sales_data = pd.read_csv('~/aiffel/data_analysis_basic/data/sales_05_8.csv')
```

```python
# 고유한 ID 수 확인
len(sales_data['ID'].unique())
```

### 2. RFM Score

```python
# 계산을 위한 전처리
sales_data['Now'] = '2022-06-10' # Now 컬럼 생성
sales_data['Now'] = pd.to_datetime(sales_data['Now']) # Now 컬럼 값을 datetime으로 뱐환
sales_data['Time_stamp'] = pd.to_datetime(sales_data['Time_stamp']) # Time_stamp 컬럼 값을 datetime으로 변환
sales_data = sales_data[sales_data['Status']=='purchase'] # Status 컬럼 값이 Purchace인 행 추출
sales_data['Interval'] = sales_data['Now'] - sales_data['Time_stamp'] # 기준 일자 - 최근 접속 일자
```

```python
# RFM_data 변수에 Recency, Frequency, Monetary 집계
RFM_data = sales_data.groupby(['ID']).agg({'Interval':'min', 'ID':'count', 'Payment':'sum'})
RFM_data.columns = ['Recency', 'Frequency', 'Monetary'] # 컬럼 값 변경
RFM_data['Recency'] = RFM_data['Recency'].dt.days # 자료형 변경
```

**A. Recency**

```python
# 기술통계량 출력
RFM_data.describe().T.round(2)
```

```python
temp_var = RFM_data.sort_values(by=['Recency'], ascending=False) # 내림차순 정렬
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# 히스토그램
ax1 = sns.histplot(data=temp_var['Recency'], ax=ax1)
ax1.set_title('Recency Histogram')

# 박스플롯
ax2 = sns.boxplot(data=temp_var, x='Recency', ax=ax2)
ax2.set_title('Recency Box plot')
plt.show()
```

* 160일 전을 기준으로 결제 수가 꾸준히 증가하다가 최근 들어 급격하게 감소함을 알 수 있음
* 이상치는 없으며 약 40일~100일 전 사이에 결제 수가 많았음

**B. Frequency**

```python
temp_var = RFM_data.sort_values(by=['Frequency'], ascending=False) # 내림차순 정렬
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# 히스토그램
ax1 = sns.histplot(data=temp_var['Frequency'], ax=ax1)
ax1.set_title('Frequency Histogram')

# 박스플롯
ax2 = sns.boxplot(data=temp_var, x='Frequency', ax=ax2)
ax2.set_title('Frequency Box plot')
plt.show()
```

* 결제 5회 이후 추가 결제를 하는 사람이 크게 줄어든 걸 확인할 수 있음
* 10번 이상의 결제는 모두 이상치로 판단될 정도로 보통 결제 횟수가 적은 걸 확인할 수 있음

**C. Monetary**

```python
temp_var = RFM_data.sort_values(by=['Monetary'], ascending=False) # 내림차순 정렬
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# 히스토그램
ax1 = sns.histplot(data=temp_var['Monetary'], ax=ax1)
ax1.set_title('Monetary Histogram')

# 박스플롯
ax2 = sns.boxplot(data=temp_var, x='Monetary', ax=ax2)
ax2.set_title('Monetary Box plot')
plt.show()
```

* 소액 결제의 수가 압도적으로 많으며 1만원 이상 결제 수는 매우 적은 걸 확인할 수 있음
* 대부분의 결제는 약 7천원 안으로 이루어지며, 약 1만 3천원 이상의 결제는 이상치로 구분되고 있음

### 3. RFM Score 기반 VIP 선정
**A. Class 분류**

```python
# 각 항목의 score 컬럼 생성(선형보간법)
RFM_data['R_score'] = np.interp(RFM_data['Recency'], (RFM_data['Recency'].min(), RFM_data['Recency'].max()), (100,1)).round(2) # R_score
RFM_data['F_score'] = np.interp(RFM_data['Frequency'], (RFM_data['Frequency'].min(), RFM_data['Frequency'].max()), (1,100)).round(2) # F_score
RFM_data['M_score'] = np.interp(RFM_data['Monetary'], (RFM_data['Monetary'].min(), RFM_data['Monetary'].max()), (1,100)).round(2) # M_score
```

```python
RFM_data['RFM_score'] = ((RFM_data['R_score'] + RFM_data['F_score'] + RFM_data['M_score'])/3).round(2) # 평균값 계산
RFM_data.sort_values(by=['RFM_score'], ascending=False, inplace=True)
RFM_data['Class'] = pd.qcut(RFM_data['RFM_score'], q=[0, 0.4, 0.65, 0.85, 0.99, 1], labels=[5, 4, 3, 2, 1]) # 등급 구분
RFM_data.reset_index(inplace=True) # 인덱스 재정렬
```

**B. 분류 검증**

```python
temp_var = RFM_data['ID'].groupby(RFM_data['Class']).count() # ID 값을 Class 컬럼 기준으로 집계

# 파이그래프 시각화
plt.pie(temp_var, labels=temp_var.index, autopct='%1.1f%%')
plt.title('Class Proportion')
plt.show()
```

* 설정한 대로 잘 구분되었음을 알 수 있음

**C. Class별 Score**

```python
# Class 컬럼을 기준으로 각 항목의 평균갑 집계
temp_var = RFM_data.groupby('Class').agg({'Recency':'mean', 'Frequency':'mean', 'Monetary':'mean'}).round(1)
temp_var.T
```

**D. Class별 Score 시각화**

```python
# MultiIndex를 NormalIndex로 변환
temp_var.index = temp_var.index.tolist()
temp_var.reset_index(inplace=True)
temp_var.rename(columns = {'index':'Class'}, inplace=True) # 인덱스명 Class로 변경
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(11, 4))

# Recency
ax1 = sns.barplot(data=temp_var, x='Class', y='Recency', ax=ax1)
containers = ax1.containers[0]
ax1.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax1.set_title('Frequency by Class')

# Frequency
ax2 = sns.barplot(data=temp_var, x='Class', y='Frequency', ax=ax2)
containers = ax2.containers[0]
ax2.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax2.set_title('Frequency by Class')

# Monetary
ax3 = sns.barplot(data=temp_var, x='Class', y='Monetary', ax=ax3)
containers = ax3.containers[0]
ax3.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax3.set_title('Monetary by Class')

plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
# VIP
VIP_customer = RFM_data[RFM_data['Class']==1]
VIP_customer.reset_index(drop=True, inplace=True)
VIP_customer.head(3)
```

### 4. 3차원 시각화 자료 생성

```python
# 라이브러리 호출
from mpl_toolkits.mplot3d import Axes3D

# figure 및 ax 정의
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# 각축에 데이터 할당
x = RFM_data['R_score']
y = RFM_data['F_score']
z = RFM_data['M_score']

# 각 축에 레이블 할당
ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')

# 산포도 형태의 3차원 시각화 (색깔은 x축을 기준으로 함)
ax = ax.scatter(x, y, z, c=x)

plt.title('RFM Score of All Customers')
plt.show()
```

* 3차원 시각화 자료를 기반으로 VIP 고객 선정 뿐 아니라, 새로운 상품 출시를 통한 구매 유도 성공 가능성이 높은 고객과 게임 이탈 가능성이 높은 고객을 추가로 선정하여 각각에 대한 대안 및 전략을 수립할 수 있음

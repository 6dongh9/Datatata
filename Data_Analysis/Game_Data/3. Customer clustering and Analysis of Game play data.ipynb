# 유저의 게임 플레이 특성 분석
### 1. Problem
* 유저의 게임 플레이 유형을 파악하여 게임 발전 방향성 설정 및 이벤트 기획에 활용할 필요가 있음  

### 2. Solution
* 군집분석(Cluster Analysis)과 상관분석을 통해 VIP 유저를 정의하고 게임 플레이 유형 파악 후 운영 전략 수립

```python
# 라이브러리 호출
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
# 데이터 로드
customer_data = pd.read_csv('~/aiffel/data_analysis_basic/data/customer_for_cluster.csv')
```

### 1. VIP 선정
**A. Clustering**

```python
from sklearn.cluster import KMeans
from sklearn import preprocessing

temp_var = customer_data[['Level_score','Time_score']] # 분석에 활용할 컬럼 추출
temp_var = preprocessing.StandardScaler().fit(temp_var).transform(temp_var) # 정규화
```

```python
# 초기 모델 생성
Kmeans = KMeans(
    n_clusters=5, # 군집 개수
    init='k-means++', # 모델 초기화 방법
    n_init=10, # 모델의 초기화 횟수
    verbose=False, # 단계별 메시지 출력 여부
    max_iter=100, # 최대 반복 횟수
    random_state=1111 # 랜덤시드 고정
)

Kmeans.fit(temp_var) # 클러스터링 수행
cluster_label = Kmeans.labels_ # 군집 이름 추출
customer_data['Cluster'] = pd.DataFrame(cluster_label) # 새로운 컬럼에 저장
```

```python
temp_var = customer_data['ID'].groupby(customer_data['Cluster']).count() # 클러스터링 결과 집계

# 파이그래프 시각화
plt.pie(temp_var, labels=temp_var.index, autopct='%1.1f%%')
plt.title('Cluster Proportion')
plt.show()
```

* 클러스터별로 비슷한 비율로 구분된 것으로 보아 플레이 특성을 잘 반영했다고 보기는 어려움

**B. Clustering 결과 시각화**

```python
# 산포도
ax = sns.scatterplot(data=customer_data, x='Level_score', y='Time_score', hue='Cluster')
plt.title('Scatter Chart By Cluster')
plt.show()
```

* 1번 클러스터는 잘 묶였지만 나머지는 구분 기준이 다소 불명확해 보임

**C. Cluster별 Score 시각화**

```python
# Class 기준 평균 집계
temp_var = customer_data.groupby('Cluster').agg({'Time_score':'mean', 'Level_score':'mean', 'Payment':'mean'}).round(1)

# MultiIndex를 NormalIndex로 변환
temp_var.index = temp_var.index.tolist()
temp_var.reset_index(inplace=True)
temp_var.rename(columns = {'index':'Cluster'}, inplace=True) # 인덱스명 Cluster로 변경
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

ax1 = sns.barplot(data=temp_var, x='Cluster', y='Time_score', ax=ax1)
containers = ax1.containers[0]
ax1.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax1.set_title('Time_score by Cluster')

ax2 = sns.barplot(data=temp_var, x='Cluster', y='Level_score', ax=ax2)
containers = ax2.containers[0]
ax2.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax2.set_title('Level_score by Cluster')

ax3 = sns.barplot(data=temp_var, x='Cluster', y='Payment', ax=ax3)
containers = ax3.containers[0]
ax3.bar_label(containers, labels=[f'{x:,.0f}' for x in containers.datavalues], label_type='center')
ax3.set_title('Payment by Cluster')

plt.subplots_adjust(wspace=0.3)
plt.show()
```

### 2. 상관관계 분석

```python
# Time_score 및 게임 활동과 관련된 컬럼 추출 후 상관관계 분석
corr_data = customer_data[['Time_score', 'PK', 'BOSS', 'Gacha', 'Attack', 'Mining']].corr().round(2)

# 히트맵 시각화
sns.heatmap(data = corr_data, annot=True)
plt.title('Heatmap By Correlation')
plt.show()
```

* Time_score와 가장 높은 상관관계를 보이는 항목은 Mining(0.63)과 PK(0.32)임을 확인할 수 있음

### 3. 게임 플레이 특성 분석
* Time_socre와 Level_score가 높은 1번 클러스터와 낮은 2번 클러스터를 비교함

```python
# Class 기준 평균 집계
temp_var = customer_data.groupby('Cluster').agg({'PK':'mean', 'BOSS':'mean', 'Gacha':'mean', 'Attack':'mean', 'Mining':'mean'}).round(1)

# 시각화를 위한 전처리: 인덱스 재설정
temp_var.index = temp_var.index.tolist()
temp_var = temp_var.reset_index(drop=True)
temp_var = temp_var.rename(columns = {'index':'Cluster'})

# Radar Chart를 이용한 집계결과 시각화
from math import pi

# Radar Chart의 각 방향 개수 정의
categories=list(temp_var)
N = len(categories)

# 방향변수 개수에 따른 레이더 차트 각(angle) 정의
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# 초기 레이더 차트 생성
plt.figure(figsize=(8,8)) #차트 크기
ax = plt.subplot(111, polar=True) #sub차트 생성

# 12시 방향을 기준으로 차트 생성
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# 레이더차트에 Cluster1 표기
values = temp_var.loc[1].tolist()
values.append(values[0])
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Cluster1')
ax.fill(angles, values, 'b', alpha=0.1)

# 레이더차트에 Cluster2 표기
values = temp_var.loc[2].tolist()
values.append(values[0])
ax.plot(angles, values, linewidth=1, linestyle='solid', label='Cluster2')
ax.fill(angles, values, 'c', alpha=0.1)

# 각 변수이름 부여
plt.xticks(angles[:-1], categories)
plt.legend()
plt.show()
```

### 4. 결론
* 비지도학습인 Clustering을 수행해보고, 다양한 시각화를 통해 유저의 게임 플레이 특성을 여러 관점에서 파악할 수 있었음
* 다만, 1번 클러스터를 제외한 나머지의 경우 클러스터링이 효과적으로 수행된 것 같지는 않음
* Time_score와 Level_score 외에 Payment_score 등 새로운 속성을 포함하여 다차원 클러스터링을 시도해 본다면 어떨까 하는 생각이 들었음
